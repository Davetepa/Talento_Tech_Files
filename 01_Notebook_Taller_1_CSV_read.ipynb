3. Caracteriscas.
    DF_1 es una base de datos que tienes informaciones generales de varias personas: id, peso, altura, edad y sexo.

    DF_2 es una base de datos que tienes informaciones generales de varias personas: id, peso, altura, edad, sexo e ingresos.

    DF_3 es una base de datos que tienes informaciones generales de varias personas: id, peso, altura y edad.

```python
# Patrones

# 1. Distribución de la edad:
print("\nDistribución de la edad:")
print(df_1['age'].describe())

# 2. Distribución del peso y la altura:
print("\nDistribución del peso y la altura:")
print(df_1[['weight', 'height']].describe())

# 3. Proporción de hombres y mujeres:
print("\nProporción de hombres y mujeres:")
print(df_1['gender'].value_counts(normalize=True))

# 4. Correlación entre peso y altura:
print("\nCorrelación entre peso y altura:")
print(df_1['weight'].corr(df_1['height']))

# 5. Edad promedio por sexo:
print("\nEdad promedio por sexo:")
print(df_1.groupby('gender')['age'].mean())
```

```python
# Tendencia

if 'gender' in df_2.columns:
  df_2.groupby('gender')['income'].mean().plot(kind='bar')
```

```python
# Datos atípicos

# Calcular el rango intercuartílico (IQR) para cada columna numérica
Q1 = df_3.quantile(0.25)
Q3 = df_3.quantile(0.75)
IQR = Q3 - Q1

# Identificar los valores atípicos (outliers)
outliers = df_3[((df_3 < (Q1 - 1.5 * IQR)) | (df_3 > (Q3 + 1.5 * IQR))).any(axis=1)]

# Mostrar los valores atípicos
print("Valores atípicos en df_3:")
print(outliers)
```

```python
# Datos faltantes

# Verificar la cantidad de valores faltantes por columna en df_2
print("Valores faltantes en df_2:")
print(df_2.isnull().sum())
```
